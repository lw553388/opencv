>1. 이미지 읽어오기 / 창 닫기

```python
import cv2 as cv  

img = cv.imread('./img_data/starry_night.jpg')  
# img = cv.imread('./img_data/starry_night.jpg',cv.IMREAD_UNCHANGED)

cv.imshow("Display window", img)
cv.waitKey(0)
cv.destroyAllWindows()
```

>cv2.IMREAD_UNCHANGED  
     이미지를 읽어들일 때 알파 채널이 있는 경우 BGRA형식으로 이미지를 읽어들이도록 함.  
     이미지 처리 작업에 따라 알파 채널의 필요성이 있을 경우에 사용함.  
    tree.png(BGRA)  
      
    tree.jpeg(BGR)  
    

>* 키보드 이벤트 처리 관련 함수 : cv.waitKey([delay])  
    함수의 인자를 지정하지 않거나 0 또는 음수로 설정하면 키 입력이 있을 때까지 무한히 대기하고, 사용자가 키보드를 누르면 눌린 키의 아스키 코드를 반환하면서 함수 종료
 

```python
import cv2

img = cv2.imread('./img_data/starry_night.jpg')
cv2.imshow('Display window', img)

while True:
 key = cv2.waitKey(0)
 if key == ord('q'): # 'q' 키 입력시 종료
    break

cv2.destroyAllWindows()
```

>2. 이미지 사이즈 조절

```python
img = cv2.resize(img,(640,640))
img32 = cv2.resize(img,(320,320))
img16 = cv2.resize(img,(160,160))
imgfx = cv2.resize(img,(0,0), fx=0.5, fy=0.3)

cv2.imshow('Display window 640', img)
cv2.imshow('Display window 320', img32)
cv2.imshow('Display window 160', img16)
cv2.imshow('Display window fx', imgfx)

cv.waitKey(0)
cv.destroyAllWindows()
```

>결과  


>3. 이미지 공간 색상 변환  
    : 이미지의 색상 표현 방식을 변경

>    * cv2.cvtColor()  
        데이터 타입을 같게 유지하고 채널을 변환  
               
>    * 자주 사용되는 파라미터  
        - cv2.COLOR_BGR2GRAY  
        - cv2.COLOR_GRAY2BGR  
        - cv2.COLOR_BGR2RGB  
        - cv2.COLOR_BGR2HSV  
        - cv2..COLOR_BGR2YUV  

```python
img = cv2.resize(img,(320,320))

img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
img_yuv = cv2.cvtColor(img, cv2.COLOR_BGR2YUV)

cv2.imshow('Display', img)
cv2.imshow('GRAY_Display', img_gray)
cv2.imshow('RGB_Display', img_rgb)
cv2.imshow('HSV_Display', img_hsv)
cv2.imshow('YUV_Display', img_yuv)

cv.waitKey(0)
cv.destroyAllWindows()
```

>결과  


>4. 이미지 저장  
    cv2.imwrite(fimename,image)

```python
img = cv2.imread('./img_data/starry_night.jpg')

img = cv2.resize(img,(320,320))

img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

cv2.imwrite('./img_data/starry_night_gray.jpg',img_gray)
```
